You can load any pic to this modal to find out whether it is Darth Vader or Stormtrooper

```python
image_data =  tf.gfile.FastGFile('tf_files/star_wars/darth_vader/test.jpeg','rb').read()
```

```python
label_lines = [line.rstrip() for line in  tf.gfile.GFile('tf_files/retrained_labels_star_wars.txt')]
```

```python
with tf.gfile.FastGFile("tf_files/retrained_graph_star_wars.pb",'rb') as f:
    graph_def = tf.GraphDef()
    graph_def.ParseFromString(f.read())
    _=tf.import_graph_def(graph_def,name='')
```

```python
with tf.Session() as sess:
    softmax_tensor = sess.graph.get_tensor_by_name('final_result:0')
    predictions = sess.run(softmax_tensor,\
                             {'DecodeJpeg/contents:0':image_data})
    top_k = predictions[0].argsort()[-len(predictions[0]):][::-1]
    for node_id in top_k:
        human_string = label_lines[node_id]
        score = predictions[0][node_id]
        print('%s(score = %.5f)' %(human_string , score))
    
```
